```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df_train =pd.read_csv('tb_hotel_traintest.csv')
df_train.head(50)
```

```python
df_train.info()
```

```python
cat_cols = [col for col in df_train.columns if df_train[col].dtype == 'O']
cat_cols
```

```python
cat_df =df_train[cat_cols]
cat_df
```

```python
num_df = df_train.drop(columns=cat_cols,axis =1)
```

```python
num_df.var()
```

```python
num_df['lead_time'] = np.log(num_df['lead_time'] + 1)
num_df['agent'] = np.log(num_df['agent'] + 1)
num_df['company'] = np.log(num_df['company'] + 1)
num_df['adr'] = np.log(num_df['adr'] + 1)
```

```python
from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()
cat_df['country']=le.fit_transform(cat_df['country'])
cat_df['hotel']=le.fit_transform(cat_df['hotel'])
cat_df['meal']=le.fit_transform(cat_df['meal'])
cat_df['market_segment']=le.fit_transform(cat_df['market_segment'])
cat_df['distribution_channel']=le.fit_transform(cat_df['distribution_channel'])
cat_df['reserved_room_type']=le.fit_transform(cat_df['reserved_room_type'])
cat_df['assigned_room_type']=le.fit_transform(cat_df['assigned_room_type'])
cat_df['deposit_type']=le.fit_transform(cat_df['deposit_type'])
cat_df['customer_type']=le.fit_transform(cat_df['customer_type'])
cat_df['reservation_status_date']=le.fit_transform(cat_df['reservation_status_date'])
cat_df['arrival_date']=le.fit_transform(cat_df['arrival_date'])
```

```python
num_df['children'].fillna(0,inplace=True)
```

```python
num_df['agent'].fillna(86.45,inplace=True)
```

```python
num_df['company'].fillna(179,inplace=True)
```

```python
num_df['adr'].fillna(4.47,inplace=True)
```

```python
df1_train= pd.concat([cat_df,num_df],axis =1)
```

```python
df1_train.head()
```
